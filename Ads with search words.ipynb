# GET DATA

```python
ads = []

request = requests.get('https://graph.facebook.com/v4.0/ads_archive',params=params).json()
data = request['data']
df = pd.DataFrame.from_dict(data)

while len(data)>0:
    
    ads.append(df)
    url = request['paging']['next']
    request = requests.get(url).json()
    data = request['data']
    df = pd.DataFrame.from_dict(data)    
```
